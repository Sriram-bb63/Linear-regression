```python
import pandas as pd

df = pd.read_csv("Fish.csv")
```

```python
df.drop(df[df["Weight"] == 0].index, inplace=True)
df["Volume"] = ((df["Length1"] + df["Length2"] + df["Length3"]) / 3) * df["Height"] * df["Width"]
df.head()
```

```python
df.shape
```

```python
x1 = df["Length1"]
x2 = df["Length2"]
x3 = df["Length3"]
x4 = df["Height"]
x5 = df["Width"]
x6 = df["Volume"]

y = df["Weight"]
```

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
plt.rcParams["figure.figsize"] = (20,10)
```

# x1 Length1

```python
x1 = pd.DataFrame(x1)

m1 = LinearRegression()
m1.fit(x1, y)
print(f"""
Slope: {m1.coef_}
y intercept: {m1.intercept_}""")
p1 = m1.predict(x1)
```

```python
plt.scatter(x1, y)
plt.ylabel("Weight")
plt.xlabel("Length1")
plt.plot(x1, p1, color="red")
plt.grid()
plt.show()
```
